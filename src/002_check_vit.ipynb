```python
import timm 
import torch
```

```python
model = timm.create_model('vit_base_patch16_224', pretrained=True)
```

```python
inputs = torch.randn(1, 3, 224, 224)
outputs = model(inputs)
outputs.shape
```

## multi channel

```python
model = timm.create_model(
    'vit_base_patch16_224', 
    pretrained=True,
    num_classes=5,
    in_chans=8,
    )
```

```python
inputs = torch.randn(1, 8, 224, 224)
outputs = model(inputs)
outputs.shape
```
